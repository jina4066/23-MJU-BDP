```python
import os
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

import urllib.parse
from datetime import datetime
import time
import pyperclip
import pyautogui
import random

options = Options()
options.add_argument('user-agent=' + "user-agent header")
options.add_argument('--mute-audio')
options.add_argument("--no-sandbox")
options.add_experimental_option("detach", True)
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(service=Service(ChromeDriverManager(driver_version="119.0.6045.105").install()), options=options)

url = 'https://nid.naver.com/nidlogin.login?mode=form&url=https%3A%2F%2Fwww.naver.com'
driver.get(url) 

driver.execute_script("document.getElementsByName('id')[0].value='sebin0580'")
time.sleep(1)
driver.execute_script("document.getElementsByName('pw')[0].value='tpqls~0045'")
time.sleep(1)

login_btn = driver.find_element(By.ID,'log.login') 
login_btn.click() 
time.sleep(2)

def move_page(page):
    search_url = "https://cafe.naver.com/nyblog?iframe_url=/ArticleSearchList.nhn%3Fsearch.clubid=15424483%26search.menuid=251%26search.searchdate=2021-04-012021-06-30%26search.searchBy=1%26search.sortBy=date%26search.option=0%26userDisplay=50%26search.query=%B3%AA%C0%CC%C5%B0%26search.includeAll=%26search.exclude=%26search.include=%26search.exact=%26search.page={}".format(page)
    return search_url


# 검색 설정
# search_bar = driver.find_element(By.XPATH, '//*[@id="query"]')
# search_bar.send_keys("나이키")  # Type "나이키" into the search bar

# setting = driver.find_element(By.XPATH, '//*[@id="currentSearchDate"]')
# setting.click()

# # 기간 설정
# min = driver.find_element(By.XPATH, '//*[@id="input_1"]')
# driver.execute_script("arguments[0].value = '2020-04-01';", min)

# time.sleep(1)

# max = driver.find_element(By.XPATH, '//*[@id="input_2"]')
# driver.execute_script("arguments[0].value = '2021-06-30';", max)

# time.sleep(1)

# search = driver.find_element(By.CSS_SELECTOR, '#btn_set')
# search.click()

# search_button = driver.find_element(By.CSS_SELECTOR, '.btn-search-green')
# search_button.click()
# time.sleep(2) 

data = [] 
# 프레임 전환이 완료된 후에 페이지 소스를 가져옴

for page in range(1, 31):
    url = move_page(page)
    driver.get(url)
    driver.switch_to.frame('cafe_main')
    search_url = driver.page_source
    soup = BeautifulSoup(search_url, 'html.parser')
    articles = soup.select('div.inner_list')

    for article in articles:
        subject = article.find('a', class_='article')

        if subject is None:
            continue

        # 새 탭에서 게시글 페이지를 엽니다.
        article_url = 'https://cafe.naver.com' + subject['href']
        driver.execute_script("window.open('');")
        driver.switch_to.window(driver.window_handles[1])
        driver.get(article_url)
        time.sleep(3)  # 페이지 로딩을 기다립니다.
        driver.switch_to.frame('cafe_main')

        # 게시글 페이지에서 가격 정보를 가져옵니다.
        article_page_source = driver.page_source
        soup = BeautifulSoup(article_page_source, 'html.parser')
        price_tag = soup.select_one('.se-module.se-module-text p:nth-of-type(4)')

        if price_tag is not None:
            price_text = price_tag.text.strip()  # '【가격】$44.94 ($33.99 + 배송비)'와 같은 형태의 텍스트
            price = price_text.split(' ')[-1]  # '【가격】'을 기준으로 나누고 마지막 요소를 가져옵니다.
            data.append(price)

        # 게시글 탭을 닫고 리스트 페이지로 돌아갑니다.
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    time.sleep(2)
    
with open('21년_2분기_나이키(해외).txt', 'w', encoding='utf-8') as file:
    for line in data:
        for l in line:
            file.write(l)
        f.write('\n')
```
